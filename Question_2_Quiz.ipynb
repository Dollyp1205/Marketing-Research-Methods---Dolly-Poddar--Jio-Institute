```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set plot style for better aesthetics
sns.set(style="whitegrid")

# -----------------------------------------------
# Step 1: Load the Data
# -----------------------------------------------
# Change the file name if needed. The file should have at least a 'neighbourhood' column.
try:
    data = pd.read_csv('/content/listings_detailed (1).csv')
except FileNotFoundError:
    # If the file is not found, generate sample data for demonstration.
    print("File not found. Generating sample data for demonstration.")
    data = pd.DataFrame({
        'neighbourhood': [
            'Downtown', 'Downtown', 'Downtown', 'Uptown', 'Uptown',
            'Midtown', 'Suburb', 'Suburb', 'Suburb', 'Suburb',
            'Beachside', 'Beachside', 'Beachside', 'Downtown', 'Uptown'
        ],
        'listing_id': range(15)
    })

# Preview the data
print("Data Preview:")
display(data.head())

# -----------------------------------------------
# Step 2: Aggregate Listings by Neighbourhood
# -----------------------------------------------
# Count the number of listings per neighbourhood
neighbourhood_counts = data['neighbourhood'].value_counts().reset_index()
neighbourhood_counts.columns = ['neighbourhood', 'listing_count']

# Display the aggregated table
print("Listings Count per Neighbourhood:")
display(neighbourhood_counts)

# -----------------------------------------------
# Step 3: Visualize the Distribution
# -----------------------------------------------
plt.figure(figsize=(10, 6))
sns.barplot(x='neighbourhood', y='listing_count', data=neighbourhood_counts, palette='viridis')
plt.title('Airbnb Listings Distribution by Neighbourhood')
plt.xlabel('Neighbourhood')
plt.ylabel('Number of Listings')
plt.xticks(rotation=45)
plt.show()

# -----------------------------------------------
# Step 4: Calculate the Gini Coefficient
# -----------------------------------------------
def gini_coefficient(array):
    """Calculate the Gini coefficient of a numpy array."""
    array = np.array(array, dtype=np.float64)
    # Shift array if there are negative values
    if np.amin(array) < 0:
        array -= np.amin(array)
    array += 1e-7  # small constant to avoid division by zero
    array = np.sort(array)
    n = array.shape[0]
    index = np.arange(1, n + 1)
    gini = (np.sum((2 * index - n - 1) * array)) / (n * np.sum(array))
    return gini

# Compute the Gini coefficient for the listing counts
gini_value = gini_coefficient(neighbourhood_counts['listing_count'])
print("\nGini Coefficient for Listings Distribution: {:.3f}".format(gini_value))

# -----------------------------------------------
# Step 5: Interpret the Results
# -----------------------------------------------
if gini_value > 0.5:
    print("Interpretation: The listings are unevenly distributed, indicating that some neighbourhoods (hot areas) have a significantly higher concentration of listings.")
else:
    print("Interpretation: The listings are relatively evenly distributed across neighbourhoods.")
```
